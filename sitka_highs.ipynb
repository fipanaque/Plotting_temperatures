```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
filename = '/content/drive/MyDrive/python_exercise/chap_16/data/sitka_weather_07_2018_simple.csv'

with open(filename) as f:
  reader = csv.reader(f)
  header_row = next(reader)
  print(header_row)
  interestingrows=[row for idx, row in enumerate(reader) if idx in (0,1)]
  display(interestingrows)
```

```python
for index, column_header in enumerate(header_row):
  print(index,column_header)
```

Extracting and reading data
### Get high temperatures from this file

```python
with open(filename) as f:
  reader = csv.reader(f)
  header_row = next(reader)

  highs = []

  for row in reader:
    high = int(row[5])
    highs.append(high)

print(highs)
```

```python
#Plot High temperatures
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(highs, c='red')

#Format plot
ax.set_title("Daily high temperature, July 2018", fontsize=24)
ax.set_xlabel('',fontsize=16)
ax.set_ylabel('Temperature (F)',fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
#Get dates and high temperatures from this file
with open(filename) as f:
  reader = csv.reader(f)
  header_row = next(reader)

  dates, highs = [], []
  for row in reader:
    current_date = datetime.strptime(row[2],'%Y-%m-%d')
    high = int(row[5])
    dates.append(current_date)
    highs.append(high)

#Plot High temperatures
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red')

#Format plot
ax.set_title("Daily high temperature, July 2018", fontsize=24)
ax.set_xlabel('',fontsize=16)
fig.autofmt_xdate()
ax.set_ylabel('Temperature (F)',fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```
